```python
y=dottadatadf.kda_ratio
X=dottadatadf.drop(["user_id","hero_id","id","kda_ratio","num_wins", "base_health", "base_health_regen", "base_mana","base_mana_regen", "base_magic_resistance","base_attack_min", "base_attack_max", "base_strength", "base_agility","base_intelligence", "strength_gain", "agility_gain","intelligence_gain", "attack_range", "projectile_speed", "attack_rate","move_speed", "turn_rate", "roles"],axis=1)
```

```python
X.head()
```

```python
X.columns
```

```python
from sklearn.ensemble import RandomForestRegressor   #Random Forest
```

```python
Randfor=RandomForestRegressor(n_estimators=500)
```

```python
Randfor.fit(X,y)
```

```python
Randfor.score(X,y)
```

```python
Randforpredict=Randfor.predict(X)
```

```python
randforresidual=y-Randforpredict
```

```python
np.sqrt(np.mean((randforresidual)**2))    #logorathmic error
```

```python
importances = Randfor.feature_importances_   ## from here to see impt features
```

```python
std = np.std([tree.feature_importances_ for tree in Randfor.estimators_],
             axis=0)
```

```python
indices = np.argsort(importances)[::-1]
```

```python
# Print the feature ranking
print("Feature ranking:")
for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```
