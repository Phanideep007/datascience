```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
```

```python
telcochurn=pd.read_csv("D:/Telecom Churn Data SingTel.csv")
```

```python
telcochurn.head()
```

```python
telcochurn.tail()
```

```python
telcochurn.describe().transpose()
```

```python
## If there are any True or False, Yes or No, T or F etc in your data you need to convert them to 0 & 1
## To do that you need to use np.where
le=LabelEncoder()
```

```python
telcochurn["International Plan"]=le.fit_transform(telcochurn.iloc[:,4].values)
```

```python
telcochurn["Voice mail Plan"]=le.fit_transform(telcochurn.iloc[:,5].values)
```

```python
telcochurn.head()
```

```python
telcochurn["Churn"]=le.fit_transform(telcochurn.iloc[:,20].values)
```

```python
telcochurn.head()
```

```python
features=telcochurn.drop(['Churn','State','Area Code','Phone Number'],axis=1).columns
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
churnrf=RandomForestClassifier()
```

```python
churnrf.fit(telcochurn[features],telcochurn["Churn"])
```

```python
churnrf.score(telcochurn[features],telcochurn['Churn'])
```

```python
predictions=churnrf.predict(telcochurn[features])
```

```python
predictions
```

```python
from sklearn import metrics
```

```python
telcochurn["predictions"]=churnrf.predict(telcochurn[features])
```

```python
telcochurn.describe().transpose()
```

```python
telcochurn.head()
```

```python
rfcm=metrics.confusion_matrix(telcochurn.Churn,telcochurn.predictions)
```

```python
rfcm
```

```python
metrics.roc_curve(telcochurn.Churn, telcochurn.predictions)
```
